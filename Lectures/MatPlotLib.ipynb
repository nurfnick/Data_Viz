# Mat Plot Lib Chaos

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/refs/heads/main/Data_Sets/AB_NYC_2019.csv')
df.head()
```

```python
plt.plot(df.price)
```

```python
fig,ax = plt.subplots(figsize=(8,4))
ax.scatter(df.price, df.number_of_reviews)
ax.title.set_text('NYC Airbnb Listing Prices')
ax.set_xlabel('Listings')
ax.set_ylabel('Price')

fig.show()
```
