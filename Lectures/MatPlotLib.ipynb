```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/refs/heads/main/Data_Sets/AB_NYC_2019.csv')
df.head()
```

```python
plt.plot(df.price)
plt.savefig('NYC_Airbnb_Prices.png')
```

```python
height =4

fig,ax = plt.subplots(figsize=(8,5))
ax.scatter(df.price, df.number_of_reviews)
ax.title.set_text('NYC Airbnb Listing Prices')
ax.set_xlabel('Price')
ax.set_ylabel('Number of Reviews')
ax.set_xscale('log')
ax.set_yscale('log')
fig.show()
```

```python
def fib(n):
  if n==0 or n==1:
    return n
  else:
    return fib(n-1) + fib(n-2)
```
