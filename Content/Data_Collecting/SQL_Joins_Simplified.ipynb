```python
%%bigquery --project pic-math
SELECT *
FROM `bigquery-public-data.irs_990.irs_990_ein`
LIMIT 10
```

```python
%%bigquery --project pic-math

SELECT *
FROM `bigquery-public-data.irs_990.irs_990_ez_2012`
LIMIT 10
```

```python
%%bigquery --project pic-math

SELECT count(*)
FROM `bigquery-public-data.irs_990.irs_990_ein` as t1
INNER JOIN  `bigquery-public-data.irs_990.irs_990_ez_2012` as t2
ON t1.ein = t2.ein
LIMIT 10

```

```python
%%bigquery --project pic-math

SELECT count(*)
FROM `bigquery-public-data.irs_990.irs_990_ein` as t1
FULL OUTER JOIN  `bigquery-public-data.irs_990.irs_990_ez_2012` as t2
ON t1.ein = t2.ein
LIMIT 10
```

```python
%%bigquery --project pic-math

SELECT count(*)
FROM `bigquery-public-data.irs_990.irs_990_ein` as t1
LEFT JOIN  `bigquery-public-data.irs_990.irs_990_ez_2012` as t2
ON t1.ein = t2.ein
LIMIT 10
```

```python
%%bigquery --project pic-math

SELECT count(*)
FROM `bigquery-public-data.irs_990.irs_990_ein` as t1
RIGHT JOIN  `bigquery-public-data.irs_990.irs_990_ez_2012` as t2
ON t1.ein = t2.ein
LIMIT 10
```
