```python
pdf = pdfplumber.open("EpsteinPhoneBill.pdf")
```

```python
page0 = pdf.pages[5]

# Crop the page
page0 = page0.crop((0.0 * float(page0.width), 0.15 * float(page0.height), 0.50 * float(page0.width),0.95 * float(page0.height)))

ts = {
    "vertical_strategy": "explicit",
    "horizontal_strategy": "text",
    "explicit_vertical_lines": [per * float(page0.width) for per in [ 0.07, 0.165, 0.30,0.45,0.55, 0.65, 0.75, 0.85,1]],  # Define explicit vertical lines. What this is doing is adding a line at the 11%, 22.6%, ... of the PDF's width.
    "intersection_x_tolerance": 10,
}

# This code saves the debug visual output.
im = page0.to_image(resolution=200)
im.reset().debug_tablefinder(ts)
im.save("image.png", format="PNG")

# Extract the tables.
tables = page0.extract_tables(table_settings=ts)
for table in tables:
    print()
    for row in table:
        print(row)
```
