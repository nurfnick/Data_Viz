```python


import requests
import pandas as pa
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0'}
r = requests.get('https://en.wikipedia.org/wiki/The_Simpsons', headers=headers)
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

```python
tables = html_soup.find_all('table')
len(tables)
```

```python
from io import StringIO
import pandas as pa

df = pa.read_html(StringIO(str(tables)))[2]
df.columns = df.columns.droplevel(0).droplevel(0)

df
```

```python
df.to_csv("LeSeimpsons.csv")
```
