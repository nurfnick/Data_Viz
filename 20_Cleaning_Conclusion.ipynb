```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/bank.csv')

df.head(10)
```

```python
df.dtypes
```

```python
df.y = df.y=='yes'
```

```python
df.dtypes
```

```python
import re

re.sub(r"-"," ",df.job[6])
```

```python
df.job.apply(lambda x: re.sub(r"-"," ",x))
```

```python
df.job = df.job.apply(lambda x: re.sub(r"-"," ",x))
```

```python
import numpy as np

df.pdays.replace(-1,np.nan)
```

```python
df.pdays = df.pdays.replace(-1,np.nan)
```

```python
df.pdays = df.pdays.fillna(0)
```

```python
df.groupby('job').balance.agg(['mean','median','count','std'])
```

```python
monthConvert = {
    'jan':1,
    'feb':2,
    'mar':3,
    'apr':4,
    'may':5,
    'jun':6,
    'jul':7,
    'aug':8,
    'sep':9,
    'oct':10,
    'nov':11,
    'dec':12
}
```

```python
date = pa.to_datetime(dict(year = 2020, day = df.day, month = df.month.map(monthConvert)))
```

```python
df = pa.concat([df,pa.DataFrame( date)],axis = 1)

df.head()
```
